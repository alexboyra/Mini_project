```python
# Create a dataframe for the reviews
reviews = pd.DataFrame({
    'review': positive_reviews + negative_reviews,
    'label': ['positive'] * len(positive_reviews) + ['negative'] * len(negative_reviews)
})

# Define additional features
def count_positive_words(review):
    return len([word for word in review.split() if word in positive_words])

def count_negative_words(review):
    return len([word for word in review.split() if word in negative_words])

def contains_no(review):
    return int('no' in review.lower())

def count_pronouns(review):
    pronouns = ['i', 'me', 'my', 'you', 'your']
    return len([word for word in review.lower().split() if word in pronouns])

def contains_exclamation(review):
    return int('!' in review)

def log_review_length(review):
    return len(review.split())

def log_review_length(review):
    length = len(review.split())
    return math.log(length) if length > 0 else 0  # Handle cases with zero length

# Extract features
reviews['positive_words'] = reviews['review'].apply(count_positive_words)
reviews['negative_words'] = reviews['review'].apply(count_negative_words)
reviews['contains_no'] = reviews['review'].apply(contains_no)
reviews['pronoun_count'] = reviews['review'].apply(count_pronouns)
reviews['contains_exclamation'] = reviews['review'].apply(contains_exclamation)
reviews['log_review_length'] = reviews['review'].apply(log_review_length) # Modified to handle zero length reviews

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    reviews.drop('label', axis=1),
    reviews['label'],
    test_size=0.2,
    random_state=42
)

# Initialize a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Transform the text data into feature vectors
X_train_tfidf = vectorizer.fit_transform(X_train['review'])
X_test_tfidf = vectorizer.transform(X_test['review'])

# Train and evaluate models
models = {
    'Multinomial Naive Bayes': MultinomialNB(),
    'Support Vector Machine': SVC(),
    'Logistic Regression': LogisticRegression()
}

for name, model in models.items():
    # Train the model
    model.fit(X_train_tfidf, y_train)

    # Make predictions on the test set
    y_pred = model.predict(X_test_tfidf)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)

    print(f'{name} Accuracy: {accuracy:.4f}')
```
